Calculating the Scores and save the Scored_main_df_adjusted.csv file

```python
import shapely
import geopandas as gpd
import pandas as pd



def csv_to_gdf(csv_file, crs='EPSG:4326', **pandas_params):
    df = pd.read_csv(csv_file, **pandas_params)
    df['geometry'] = df['geometry'].apply(shapely.wkt.loads)
    gdf = gpd.GeoDataFrame(df, geometry='geometry', crs=crs)
    return gdf

def get_scores(df, dct, boundaries=['Country', 'Province', 'District', 'DS']):
    df['Country'] = 'SL'
    cols = [k for k, v in dct.items() if v != 0]

    for area in boundaries:
        norm_df = df.copy()

        for c in cols:
            norm_df[c] = norm_df[c].fillna(0)
            norm_df[c] = norm_df.groupby(area)[c].transform(lambda x: x / x.max() if x.max() != 0 else 0)
            norm_df[c] = dct[c] * norm_df[c]
        norm_df['score'] = norm_df[cols].sum(axis=1)
        norm_df['score'] = norm_df.groupby(area)['score'].transform(lambda x: x / x.max() if x.max() != 0 else 0)
        df[f'{area}_Score'] = norm_df['score'].copy()

        norm_df['rank'] = norm_df.groupby(area)['score'].rank(method='dense', ascending=False)
        df[f'{area}_Score'] = norm_df['score'].copy()
        df[f'{area}_Rank'] = norm_df['rank'].copy()

    df = df[[c for c in df.columns if c != 'geometry' and c != 'Country'] + ['geometry']]
    return df

def get_radius(df, km):
    temp = df[['geometry']].copy().to_crs('EPSG:3857')
    temp['geometry'] = [g.buffer(km * 1000) for g in temp['geometry']]
    temp = temp.to_crs('EPSG:4326')
    df['radius'] = temp['geometry']
    return df


def get_market_share(df, branches, competitors, radius):
    radius_df = get_radius(df[['hex_id', 'geometry']].copy(), radius)
    radius_df = gpd.GeoDataFrame(radius_df[['hex_id', 'radius']], geometry='radius')
    radius_df = gpd.sjoin(left_df=radius_df, right_df=branches[['Name', 'geometry']], how='left')
    radius_df = radius_df.groupby(['hex_id', 'radius'], as_index=False).count()
    radius_df = gpd.GeoDataFrame(radius_df[['hex_id', 'radius', 'Name']], geometry='radius')
    radius_df = gpd.sjoin(left_df=radius_df, right_df=competitors[['Competitor', 'geometry']], how='left')
    del radius_df['index_right']
    radius_df = radius_df.groupby(['hex_id', 'radius', 'Name'], as_index=False).count()
    radius_df['Area_Presence'] = (radius_df['Name'] / (radius_df['Name'] + radius_df['Competitor'])).round(2)
    radius_df = radius_df.set_index('hex_id')
    for c in ['Name', 'Competitor', 'Area_Presence']:
        df[c] = df['hex_id'].map(radius_df[c])
    return df



full_df = csv_to_gdf('main_df_adjusted.csv')
df = full_df.copy()
cargills = csv_to_gdf('cargills_with_hex.csv')
competitors = csv_to_gdf('competitors.csv')
radius = 2  # km

initial_weights = {
  "Population": 3,
  "Latch": 2,
  "Bank": 1,
  "Pharmacy": 1,
  "Fuel_Station": 1,
  "School": 1,
  "Supermarket": 1,
  "Bank_User_5": 2,
  "Eat_Out_5": 2,
  "Tourists": 2
}

weights=initial_weights

# MARKET SHARE
df = get_market_share(df, cargills, competitors, radius)

# SCORES
df = get_scores(df, weights)
df.to_csv('Scored_main_df_adjusted.csv', index=False)
```

Adding   'hex_id', 'Province', 'District', 'DS' to the Competitor.csv and create Competitor_with_hex.csv

```python
import pandas as pd
import geopandas as gpd
from shapely import wkt

# Load the data
def load_data():
    # Load competitor data
    competitors_df = pd.read_csv('competitors.csv')
    # Convert the geometry string to shapely Point objects
    competitors_df['geometry'] = competitors_df['geometry'].apply(wkt.loads)
    # Convert to GeoDataFrame
    competitors_gdf = gpd.GeoDataFrame(competitors_df, geometry='geometry', crs="EPSG:4326")

    # Load hex data
    hex_df = pd.read_csv('main_df_adjusted.csv')
    # Convert the geometry string to shapely Polygon objects
    hex_df['geometry'] = hex_df['geometry'].apply(wkt.loads)
    # Convert to GeoDataFrame
    hex_gdf = gpd.GeoDataFrame(hex_df, geometry='geometry', crs="EPSG:4326")

    return competitors_gdf, hex_gdf

def spatial_join():
    # Load the data
    competitors_gdf, hex_gdf = load_data()

    # Perform spatial join to find which hex_id each competitor point falls into
    # 'within' operation checks if each point falls within a polygon
    joined_gdf = gpd.sjoin(competitors_gdf, hex_gdf, how='left', predicate='within')

    # Select only the columns we want in the output
    result_df = joined_gdf[[
        'hex_id', 'Province', 'District', 'DS',
        'Name', 'Amenity', 'geometry', 'Competitor'
    ]]

    # Save to CSV
    result_df.to_csv('competitors_with_hex.csv', index=False)

    print(f"Processing complete. Found hex regions for {len(result_df)} competitor points.")
    print(f"Results saved to 'competitors_with_hex.csv'")

    # Return a preview of the results
    return result_df.head()

if __name__ == "__main__":
    result_preview = spatial_join()
    print("\nPreview of results:")
    print(result_preview)
```
